```python
# brute force for counting inversions
def bruteInv(array):
    i = 0
    l = len(array)
    count = 0
    while i < l:
      j = i+1
      while j < l:
         if array[i] > array[j]:
            count += 1 
         j += 1
      i += 1
    return count       
```

```python
# inversion algorithm by piggybacking on merge sort 
# how does the algorithm handles duplicates? b 
def sortAndCount(arr):
    l = len(arr)
    count = 0
    if l > 1: 
        mid = l // 2
        left = arr[:mid]
        right = arr[mid:]        
        count += sortAndCount(left)
        count += sortAndCount(right)
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                arr[k] = left[i]
                i += 1 
            else:
                arr[k] = right[j]
                count += len(left)-i 
                j += 1
            k += 1
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
    return count
```

```python
import random
import time  
import pandas as pd 

def genRandArr(k):
    r_arr = [random.randint(1,10000) for _ in range(k)]
    return r_arr

params = [{'k':50},
          {'k':500},
          {'k':5000},
          {'k':50000}

          ]

result = []

for p in params:
    r_array = genRandArr(**p)

    start = time.time()
    b_count = bruteInv(r_array)
    end = time.time()
    brute_time = end - start

    start = time.time()
    m_count = sortAndCount(r_array)
    end = time.time()
    merge_time = end -start
    p.update({'brute_time':brute_time,'merge_time':merge_time,'brute_count':b_count,'merge_count':m_count})
    result.append(p)

results = pd.DataFrame(result)

print(results)
```
