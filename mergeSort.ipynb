```python
def mergeSort(arr):
    l = len(arr)
    if l>1:
        mid = l//2
        left = arr[:mid]
        right= arr[mid:]
        
        mergeSort(left)
        mergeSort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1
        
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1 
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1 
```

```python
def mergeSubRoutine(a,b):
    merged = []
    i = j = 0
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            merged.append(a[i])
            i += 1
        else:
            merged.append(b[j])
            j += 1
    while i < len(a):
        merged.append(a[i])
        i += 1
    while j < len(b):
        merged.append(b[j])
        j += 1
    return merged
```

```python
def mergeKSortedArrays(arrays):
    while len(arrays) > 1:
        merged_arrays = []        
        for i in range(0, len(arrays), 2):
            if i + 1 < len(arrays):
                merged = mergeSubRoutine(arrays[i], arrays[i + 1])
                merged_arrays.append(merged)
            else:
                merged_arrays.append(arrays[i])
        arrays = merged_arrays
    return arrays[0]
```

```python
import random
import time
from functools import reduce
def runningTime(k=2000,n=20):
    list_of_sorted_arrays = []
    for i in range(k):
        arr = [random.randint(1, 10000) for _ in range(n)]
        mergeSort(arr)
        list_of_sorted_arrays.append(arr)
    start_time_naive = time.time()
    mergeKsortedArrayNaive = reduce(mergeSubRoutine, list_of_sorted_arrays)
    end_time_naive = time.time()

    start_time_pairs = time.time()
    mergeKsortedArrayPairs = mergeKSortedArrays(list_of_sorted_arrays)
    end_time_pairs = time.time()

    return({'naiveTime': end_time_naive - start_time_naive,
            'pairsTime': end_time_pairs - start_time_pairs})
```

```python
import pandas as pd
test_cases = [
    {'k': 100, 'n': 20},
    {'k': 500, 'n': 20},
    {'k': 1000, 'n': 20},
    {'k': 2000, 'n': 20},
    {'k': 2000, 'n': 50},
    {'k': 3000, 'n': 50},
    {'k': 4000, 'n': 50},
    {'k': 5000, 'n': 50}
]
results = []
for params in test_cases:
    result = runningTime(**params)
    result.update(params) 
    results.append(result)

df_results = pd.DataFrame(results)
print(df_results) # using native tim sort for sorting individual arrays
```
