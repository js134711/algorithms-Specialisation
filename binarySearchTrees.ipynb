```python
class Node:
    def __init__(self,key=None,value=None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None           
```

```python
def findSplitNode(tree,key1,key2):
    assert key1 <= key2 
    v = tree
    while v.left is not None and v.right is not None and (key2 <= v.key or key1 > v.key):
        if key2 <= v.key:
            v = v.left
        else:
            v = v.right
    return v

def reportSubtree(node,out = []):
    if node.left is None and node.right is None:
        out.append((node.key,node.value))
    else:
        reportSubtree(node.left,out)
        reportSubtree(node.right,out)

def rangeQuery(tree,x1,x2):
    assert x1<=x2
    out = []
    vsplit = findSplitNode(tree,x1,x2)
    if vsplit.left is None and vsplit.right is None:
        if x1<=vsplit.key<=x2:
            out.append((vsplit.key,vsplit.value))
            return out
    v = vsplit.left
    while v.left is not None and v.right is not None:
        if x1 <= v.key:
            reportSubtree(v.right,out)
            v = v.left
        else:
            v = v.right
    if x1<=v.key<=x2:
        out.append((v.key,v.value))
    v = vsplit.right
    while v.left is not None and v.right is not None:
        if x2 >= v.key:
            reportSubtree(v.left,out)
            v = v.right
        else:
            v = v.left 
    if x1<=v.key<=x2:
        out.append((v.key,v.value))
    return out 
```

```python
# building a balanced binary search tree 
# internal node contains maximum of left subtree
test = {1.1:'a', 2.2:'b', 3.3:'c', 4.4:'d', 5.5:'e', 6.6:'f', 7.7:'g', 8.8:'h', 9.9:'i', 10.10:'j'}

nodes = []

for key,value in test.items():
    nodes.append(Node(key,value))

while len(nodes) > 1: 
    new_level = []
    for i in range(0,len(nodes),2):
        if i+1 < len(nodes):
            left,right = nodes[i],nodes[i+1]
            parent = Node()
            parent.left = left 
            parent.right = right
            temp = left
            while temp.right is not None:
                temp = temp.right

            parent.key = temp.key 
            new_level.append(parent)
        else:
            new_level.append(nodes[i])
    nodes = new_level

root = nodes[0]
```

```python
rangeQuery(root,0,5)
```
